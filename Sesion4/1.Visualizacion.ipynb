```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,4,6,7,8]
plt.plot(x,y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Grafico de linea')
plt.show()
```

```python
#Grafico de dispersion
plt.scatter(x,y)
plt.title('Grafico de dispersion')
```

```python
import numpy as np
datos=np.random.randn(1000)

plt.hist(datos,color='skyblue',edgecolor='black',bins=30)
plt.title('Histograma')
plt.show()
plt.savefig('histograma.jpg')
```

```python
#Diagrama de barras
categorias=['A','B','C','D']
valores=[24,56,23,90]
plt.bar(categorias,valores,color='green')
plt.show()
```
